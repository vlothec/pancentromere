```python
Chr1['Chr']='Chr1'
Chr1_new=Chr1[["Chr", "T_sfs"]]
Chr1_new.head(n=5)

Chr2['Chr']='Chr2'
Chr2_new=Chr2[["Chr", "T_sfs"]]
Chr2_new.head(n=5)

Chr3['Chr']='Chr3'
Chr3_new=Chr3[["Chr", "T_sfs"]]
Chr3_new.head(n=5)

Chr4['Chr']='Chr4'
Chr4_new=Chr4[["Chr", "T_sfs"]]
Chr4_new.head(n=5)

Chr5['Chr']='Chr5'
Chr5_new=Chr5[["Chr", "T_sfs"]]
Chr5_new.head(n=5)

ChrAll_new=pd.concat([Chr1_new, Chr2_new, Chr3_new, Chr4_new, Chr5_new], ignore_index=True)
ChrAll_new=pd.melt(ChrAll_new, id_vars="Chr", value_name="T_sfs")
```
